```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Leer los datos de archivo csv, typed_uanl.csv con el URL
url = "C:/Users/PC/Documents/GitHub/GitFlow-en-Github/ML003/Practica 4/Practica/RentasBicicletasSEUL.csv"
df = pd.read_csv(url)
```

```python
# Mostrar las primeras filas del DataFrame
print(df.head())
```

```python
# variables  relevantes para la clusterizacion 
#"Rented Bike Count", "Hour", "Temperature(°C)", "Humidity(%)",
```

```python
# Seleccionar características para la clusterización
features = ["Rented Bike Count", "Hour", "Temperature(°C)", "Humidity(%)", "Wind speed (m/s)", "Visibility (10m)"]
X = df[features]

# Normalizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Aplicar K-means
kmeans = KMeans(n_clusters=3, random_state=0)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# Visualizar los clusters
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=df['Cluster'], cmap='viridis')
plt.xlabel("Rented Bike Count (scaled)")
plt.ylabel("Hour (scaled)")
plt.title("Clusterización de Renta de Bicicletas")
plt.savefig("C:/Users/PC/Documents/GitHub/GitFlow-en-Github/ML003/Practica 4/Practica/GraficaCluster.png")
plt.show()
plt.close()
```
